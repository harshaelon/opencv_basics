```python
img3=cv2.imread(r"C:\Users\HARSHA\OneDrive\Desktop\Important\Clean Work Place LinkedIn Banner.png")
```

```python
#here we are blurring the image.


img4=cv2.cvtColor(img3,cv2.COLOR_BGR2HSV)

figure_size=10

result=cv2.blur(img4,(figure_size,figure_size))

plt.figure(figsize=(11,6))


# Original image
plt.subplot(121)
plt.imshow(cv2.cvtColor(img4, cv2.COLOR_HSV2RGB))
plt.title("Original")
plt.xticks([]), plt.yticks([])

# Blurred image
plt.subplot(122)
plt.imshow(cv2.cvtColor(result, cv2.COLOR_HSV2RGB))
plt.title("Blurred")
plt.xticks([]), plt.yticks([])

plt.show()
```

```python
# rotating the image 

im1=cv2.imread(r"C:\Users\HARSHA\OneDrive\Desktop\Celebrating STEVE.png")

height,width = im1.shape[:2]

m=cv2.getRotationMatrix2D((width/2,height/2),360,1)

rotate_image=cv2.warpAffine(im1,m,(height,width))

plt.imshow(rotate_image)
```
